```python
def nmf_cluster(mat,n_components=9,init = 'random',algo='bpp'):#, 
    
    #from sklearn.datasets import fetch_20newsgroups
    from time import time
    n_samples = mat.shape[0]
    n_features =mat.shape[1]
    #n_components = 12
    #n_top_words = 20
    #batch_size = 28
    #init = "nndsvda"

    #sp_mat=sp.csr_matrix(mat)
    

    # Fit the NMF model:frobenius
    print(
    "Fitting the NMF model (Frobenius norm) with tf-idf features, "
    "n_samples=%d and n_features=%d..." % (n_samples, n_features)
    )
    t0 = time()
    from nmf import run_nmf
    W,H,err = run_nmf(mat, n_components,use_gpu=True,beta_loss=2.0, init=init, algo=algo )#,verose=False,n_jobs
    #print(err)
    print("done in %0.3fs." % (time() - t0))
    #scale_smc_nmf.index
    return W,H
```
