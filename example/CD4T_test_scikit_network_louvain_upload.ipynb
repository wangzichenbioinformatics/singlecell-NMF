```python

from sklearn.decomposition import PCA
PCA_data = PCA(n_components=10).fit_transform(df0.values)
from sklearn.neighbors import kneighbors_graph
A = kneighbors_graph(PCA_data, 10, mode='connectivity', include_self=False)
A.toarray().shape
```

```python
A.toarray()
```

```python
np.max(A.toarray())
```

```python
from sknetwork.clustering import Louvain, get_modularity
louvain = Louvain(resolution=0.2)
#labels = louvain.fit_transform(A)
#labels
labels = louvain.fit_predict(A)
labels
```

```python
np.max(labels)
```

```python
pd.DataFrame(labels,columns=['louvain_labels'],index=df0.index)
```
